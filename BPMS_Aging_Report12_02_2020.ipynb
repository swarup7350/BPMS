```
#Adding Service Name column
Techliregiulb=Techliregiulb.assign(Service_Name='Technical License Registration')[['Service_Name'] + Techliregiulb.columns.tolist()]
Techliregiulb.info()
```

```
#Adding onemore column Proposal Code
Techliregiulb=Techliregiulb.assign(Proposal_Code= np.arange(len(Techliregiulb)))[['Proposal_Code'] + Techliregiulb.columns.tolist()]
Techliregiulb.info()
Techliregiulb
```

```
frames = [ulb1, Techliregiulb]
Pendingfromulb = pd.concat(frames)
Pendingfromulb.info()
Pendingfromulb
```

```
mearge_op = pd.merge(ulbm,
                 Pendingfromulb[['Region', 'District', 'ULB Name', 'Proposal_Code', 'Service_Name', 'Registration Date', 'Days', 'Reason']],
                 on=['Region', 'District', 'ULB Name'],
                 how='left')

mearge_op.info()
mearge_op
```

```
distwise_reg = mearge_op.groupby("Days").agg({"Proposal_Code": pd.Series.nunique})
Dayswisereg=distwise_reg.sort_values(by=['Proposal_Code'])
Daysgroup_df = pd.DataFrame(columns=['Daysgroup', 'Dayscount'])
Dayswisereg=Dayswisereg.sort_values(by=['Days'])
nmaxagemore45 =0
nmaxagemore30to45 =0
nmaxagemore15to30 =0
nmaxagemore7to15 =0
nmaxagemore0to7 =0
for index1, row in Dayswisereg.iterrows():     
  if index1>45:
    nmaxagemore45=nmaxagemore45+row['Proposal_Code']
  elif (index1>=30 and index1<45):
    nmaxagemore30to45=nmaxagemore30to45+row['Proposal_Code']
  elif (index1>=15 and index1<30):
    nmaxagemore15to30=nmaxagemore15to30+row['Proposal_Code']
  elif (index1>=7 and index1<15):
    nmaxagemore7to15=nmaxagemore7to15+row['Proposal_Code']
  elif (index1>=0 and index1<7):
    nmaxagemore0to7=nmaxagemore0to7+row['Proposal_Code']
Daysgroup_df = Daysgroup_df.append({'Daysgroup': 'Days Group Above 45', 'Dayscount':nmaxagemore45}, ignore_index=True)
Daysgroup_df = Daysgroup_df.append({'Daysgroup': 'Days Group 31 to 45', 'Dayscount':nmaxagemore30to45}, ignore_index=True)
Daysgroup_df = Daysgroup_df.append({'Daysgroup': 'Days Group 16 to 30', 'Dayscount':nmaxagemore15to30}, ignore_index=True)
Daysgroup_df = Daysgroup_df.append({'Daysgroup': 'Days Group 8 to 15', 'Dayscount':nmaxagemore7to15}, ignore_index=True)
Daysgroup_df = Daysgroup_df.append({'Daysgroup': 'Days Group 0 to 7', 'Dayscount':nmaxagemore0to7}, ignore_index=True)
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
objects = Daysgroup_df['Daysgroup']
y_pos = np.arange(len(objects))
performance = Daysgroup_df['Dayscount']
performance

# plt.bar(y_pos, performance, align='center', alpha=0.5)
# plt.xticks(y_pos, objects)
# plt.ylabel('Proposal Code')
# plt.xlabel('Days Group')
# plt.xticks(rotation=90)
# plt.title('Aging Report')
# plt.show()
```
