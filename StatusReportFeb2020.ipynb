#**Process For ULB Pendancy Report**

##**Renaming Columns From Both ULB Files for further process**

###**Renaming Column From ULB Master File**

```
ulbm.rename(columns = {'Sr.No.':'Sr No'}, inplace = True)
ulbm1=ulbm
```

```
ulbm1.info()
```

###**Replace Column Name From "Pendancy Report ULB" For matching Column with other file column name**

```
#Replace Column Name From 1st File For matching Column names with other file column name
ulbpending.rename(columns = {'Service Name':'Service_Name', 'Proposal Code':'Proposal_Code','Sr No.':'Sr No',
                              'Reasons':'Reason','Application Received Date':'Registration Date', 
                        'Pending By':'Pending From Name'}, inplace = True)
ulbpending1=ulbpending
```

##**File Mearging**

###**Mearging "Pendancy Report ULB" with "Pending from Tech Authority"**

```
#Mearging ULB Proposal Pending File
frames = [ulbpending1, techulbpending2]
Pendingfromulb = pd.concat(frames)
Pendingfromulb.info()
```

###**Mearing "Pendingfromulb"with "ULB Type Master"**

```
#Merging Pendingfromulb with ULB Master ulbm1
mearge_op = pd.merge(ulbm1,
                 Pendingfromulb[['Region', 'District', 'ULB Name', 'Proposal_Code', 'Service_Name', 'Registration Date', 'Days', 'Reason']],
                 on=['Region', 'District', 'ULB Name'])
                #  how='left')

mearge_op.info()
```

```
#Info of "Reson" column to get resons details
mearge2 = mearge_op.groupby(['Reason']).agg({"Proposal_Code": pd.Series.nunique}).apply(list)
opinfo_1 = mearge2.reset_index()
opinfo_1
```

```
#Appling pivot table on new ULB Merge(mearge_op)
# pivot_1=pd.pivot_table(mearge_op,index=["Region","District","ULB Name", "ULB Type"],values=["Proposal_Code"],
#                columns=["Service_Name", "Reason"],aggfunc=[len],fill_value=0,margins=True,dropna=True)
# pivot_1
```

###**Renaming some resons from final mearge**

```
mearge_op.replace({'Pending from ULB and 1. NOC pending from Arogya department':'NOC Pending',
                   'Pending from ULB and 1. NOC pending from Bandhkam Vibhag':'NOC Pending',
                   'Pending from ULB and 1. NOC pending from City Sanitation department2. NOC pending from Garden Department':'NOC Pending',
                   'Pending from ULB and 1. NOC pending from FIRE DEPARMENT':'NOC Pending',
                   'Pending from ULB and 1. NOC pending from Fire Deapartment':'NOC Pending',
                   'Site Inspection Pending and 1. NOC pending from Fire Department2. NOC pending from Property Tax Department3. NOC pending from Water Tax Department':'NOC Pending',
                   'Site Inspection Pending and 1. NOC pending from Water Department2. NOC pending from Assessment Department':'NOC Pending',
                   'Site Inspection Pending':'Site Visit Pending',
                   'Sitevisit Pending':'Site Visit Pending',
                   'survior Sitevisit Pending':'Site Visit Pending',
                   'Authority pending and 1. NOC pending from Arogya department':'NOC Pending',
                   'Pending from ULB and':'NOC Pending'}, inplace = True)
mearge_op2=mearge_op
```

```
#Info of final mearging file after replacing resons
mearge3 = mearge_op2.groupby(['Reason']).agg({"Proposal_Code": pd.Series.nunique}).apply(list)
opinfo_2 = mearge3.reset_index()
opinfo_2
```

#**Final Output of ULB Proposal Pendancy Report**

```
pivot_2=pd.pivot_table(mearge_op2,index=["Region","District","ULB Name", "ULB Type"],values=["Proposal_Code"],
               columns=["Service_Name", "Reason"],aggfunc=[len],fill_value=0,margins=True,dropna=True)
pivot_2
```
